```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
```

```python
# 이미지 파일 경로
image_path = '/root/ai/dataset/classification_aug_0816/train/1/Type1_iter3_jpg.rf.120ab2fecf2da5c5636192b9a1f1e4af_mask_aug_23.png'

# 이미지 로드
image = Image.open(image_path).convert("L")  # "L" 모드는 이미지를 흑백으로 변환

# 이미지 전처리 및 텐서 변환
transform = transforms.Compose([
    transforms.ToTensor()  # 이미지를 텐서로 변환
])
image_tensor = transform(image)

unique_values = torch.unique(image_tensor)
print("Unique values in the tensor:", unique_values)


# 텐서 시각화
plt.imshow(image_tensor.squeeze(), cmap='gray')
plt.title("Image Tensor")
plt.axis('off')
plt.show()
# 텐서 정보 출력
print("Tensor shape:", image_tensor.shape)
print("Tensor dtype:", image_tensor.dtype)
print("Tensor values:", image_tensor)

# 텐서 내의 고유한 값들 확인
unique_values = torch.unique(image_tensor)
print("Unique values in the tensor:", unique_values)
```

```python
from PIL import Image
import numpy as np
import torch
import matplotlib.pyplot as plt

# 이미지 파일 경로
image_path = '/root/ai/dataset/poo_1-299/train/2_jpeg.rf.5c882c80accf88c75042a47a351e828c_mask.png'
# image_path = '/root/ai/dataset/poo_1-299/train/1_jpeg.rf.e8eb49e7b1de5f5268b562e70706d55b_mask.png'
# 이미지 로드
image = Image.open(image_path).convert("L")  # "L" 모드는 이미지를 흑백으로 변환

# 이미지 전처리 및 텐서 변환
image_np = np.array(image)
image_tensor = torch.tensor(image_np, dtype=torch.float32)  # 텐서로 변환

# 텐서 내의 고유한 값들 확인
unique_values = torch.unique(image_tensor)
print("Unique values in the tensor:", unique_values)

# 텐서 시각화
plt.imshow(image_tensor, cmap='gray')
plt.title("Image Tensor")
plt.axis('off')
plt.show()

# 텐서 정보 출력
print("Tensor shape:", image_tensor.shape)
print("Tensor dtype:", image_tensor.dtype)
print("Tensor values:", image_tensor)
```
